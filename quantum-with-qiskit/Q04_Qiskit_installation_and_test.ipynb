### Execute an example quantum program


1) Create a quantum circuit

```python
# import the objects from qiskit
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit_aer import AerSimulator
from random import randrange

# create a quantum circuit and its register objects

qreg = QuantumRegister(2) # quantum register with two quantum bits
creg = ClassicalRegister(2) # classical register with two classical bit
circuit = QuantumCircuit(qreg,creg) # quantum circuit composed by a quantum register and a classical register

# apply a Hadamard gate to the first qubit, creates super position
# Qubit 0 is in a 50/50 superposition: half chance of being 0, half chance of being 1.
circuit.h(qreg[0])

# set the second qubit to state |1>
# Qubit 1 has been flipped to a definite 1.
circuit.x(qreg[1])

print(circuit.draw(output='text'))
# q4_0: ┤ H ├    [ H ]: This shows the Hadamard gate hitting the first qubit, splitting it into a 50/50 superposition.
#       ├───┤ 
# q4_1: ┤ X ├    [ X ]: This shows the NOT gate hitting the second qubit at the same time, flipping it from 0 to 1.   
#       └───┘
# c4: 2/═════    Right now, this line is empty because we drew the circuit before the measurement gate (circuit.measure) was added



# apply CNOT(first_qubit,second_qubit)
# create entaglement by applying A Controlled-NOT (CNOT) gate is applied. 
# The first qubit (qreg[0]) acts as the control, and the second qubit (qreg[1]) acts as the target.
# Right after those two gates, it creates an interaction between them.

# When the CNOT (cx) gate steps in, Qubit 0 becomes the Control and Qubit 1 becomes the Target. 
# The CNOT applies a rule across the superposition:
# Branch A (If Qubit 0 is 0): The CNOT does nothing. Qubit 1 stays 1. This results in the state 01.
# Branch B (If Qubit 0 is 1): The CNOT triggers! It flips Qubit 1 from 1 to 0. This results in the state 10.

circuit.cx(qreg[0],qreg[1])

# measure the both qubits
# Both qubits are measured, and their quantum states are collapsed into classical data stored in creg.
# Because of the entanglement created in Step 4, the qubits are perfectly correlated in an opposite manner. 
# If you run this circuit on a simulator, you will get one of two outcomes with a 50/50 chance:
qbit_measure = circuit.measure(qreg,creg)

print("The execution of the cell was completed, and the circuit was created :)")
```

2) Draw the circuit

_Run the cell once more if the figure is not shown_

```python
# draw circuit 
circuit.draw(output='text')
# circuit.draw(output='mpl')
# the output will be a "matplotlib.Figure" object
```

3) Execute the circuit 1024 times in the local simulator and print the observed the outcomes

```python
## execute the circuit 1024 times
job = AerSimulator().run(circuit,shots=1024)
# get the result
counts = job.result().get_counts(circuit)
print(counts)
```

### Complete test

Here we put all methods and objects of Qiskit used througout this tutorial (see [Qiskit: Quick Reference](Q01_Qiskit_Reference.ipynb) for a complete list).

By running the cell below, you should be able to test whether your Qiskit installation is ready for our tutorial.

_It was tested with Qiskit 1.0.0 on April 12, 2024._

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit_aer import AerSimulator, UnitarySimulator, StatevectorSimulator
from math import pi

# A quantum circuit is composed by a quantum and a classical register in Qiskit
q = QuantumRegister(5,"q") # this quantum register has 5 qubits
c = ClassicalRegister(5,"c") # this classical register has 5 classical bits

qc = QuantumCircuit(q,c) # a quantum circuit with quantum and classical registers q and c

qc.x(q[2]) # apply x-gate to the qubit "q[2]" of the quantum circuit "qc"
qc.z(q[4]) # apply z-gate to the qubit "q[4]" of the quantum circuit "qc"
qc.h(q[1]) # apply h-gate to the qubit "q[1]" of the quantum circuit "qc"

a = pi/7
qc.ry(2*a,q[3]) # apply ry-gate with the specified angle to the qubit "q[3]" of the quantum circuit "qc"

qc.cx(q[1],q[0]) # apply cx-gate to the qubits "q[1]" and "q[0]" of the quantum circuit "qc"

qc.cu(2*a,0,0,0,q[1],q[0]) # apply controlled rotation operator with the specified angle 
                          #  to the qubits "q[1]" and "q[0]" of the quantum circuit "qc"

qc.ccx(q[2],q[1],q[0]) # apply ccx-gate to the qubits "q[2]", "q[1]", and "q[0]" of the quantum circuit "qc"

qc.barrier()

qc.measure(q[0],c[3]) # measure q[0] and write the output on c[3]
qc.measure(q[1],c[1]) # measure q[1] and write the output on c[1]
qc.measure(q[2],c[4]) # measure q[2] and write the output on c[4]
qc.measure(q[3],c[0]) # measure q[3] and write the output on c[0]
qc.measure(q[4],c[2]) # measure q[4] and write the output on c[2]

display(qc.draw())

qc.barrier()
## TODO:
qc.h(q[0]).c_if(c,5)
qc.x(q[0]).c_if(c,3)
qc.z(q[0]).c_if(c,1)

for i in range(5):
    if i % 2 ==0: qc.h(q[i])
    else: qc.x(q[i])
        
b = 3*pi/11
qc.cu(2*b,0,0,0,q[1],q[0])

display(qc.draw(output='mpl',reverse_bits=True))

job = AerSimulator().run(qc,shots=1000)
counts = job.result().get_counts(qc)   
print("measurement outcomes are:",counts)
print()

q = QuantumRegister(2,"q")
c = ClassicalRegister(2,"c")

qc = QuantumCircuit(q,c)

qc.h(q[0])
qc.ry(a,q[0])
qc.x(q[1])
qc.cx(q[0],q[1])

job = UnitarySimulator().run(qc,shots=1,optimization_level=0)
precision = 3
current_unitary=job.result().get_unitary(qc,precision).data

print("unitary matrix is")
print(current_unitary)
print()

job = StatevectorSimulator().run(qc,optimization_level=0)
precision = 3
current_quantum_state=job.result().get_statevector(qc,precision).data
print("quantum state is")
for val in current_quantum_state:
    print(str(val.real)+"+"+str(val.imag)+"j")

print()
print("All looks fine :-)")
```

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit_aer import AerSimulator
from math import pi

# --- PART 1: 5-Qubit Circuit with Measurements and Conditions ---

# A quantum circuit is composed by a quantum and a classical register in Qiskit
q = QuantumRegister(5, "q") # this quantum register has 5 qubits
c = ClassicalRegister(5, "c") # this classical register has 5 classical bits

qc = QuantumCircuit(q, c) # a quantum circuit with quantum and classical registers q and c

qc.x(q[2]) # apply x-gate to the qubit "q[2]" of the quantum circuit "qc"
qc.z(q[4]) # apply z-gate to the qubit "q[4]" of the quantum circuit "qc"
qc.h(q[1]) # apply h-gate to the qubit "q[1]" of the quantum circuit "qc"

a = pi/7
qc.ry(2*a, q[3]) # apply ry-gate with the specified angle to the qubit "q[3]" of the quantum circuit "qc"

qc.cx(q[1], q[0]) # apply cx-gate to the qubits "q[1]" and "q[0]" of the quantum circuit "qc"

qc.cu(2*a, 0, 0, 0, q[1], q[0]) # apply controlled rotation operator with the specified angle 
                               # to the qubits "q[1]" and "q[0]" of the quantum circuit "qc"

qc.ccx(q[2], q[1], q[0]) # apply ccx-gate to the qubits "q[2]", "q[1]", and "q[0]" of the quantum circuit "qc"

qc.barrier()

qc.measure(q[0], c[3]) # measure q[0] and write the output on c[3]
qc.measure(q[1], c[1]) # measure q[1] and write the output on c[1]
qc.measure(q[2], c[4]) # measure q[2] and write the output on c[4]
qc.measure(q[3], c[0]) # measure q[3] and write the output on c[0]
qc.measure(q[4], c[2]) # measure q[4] and write the output on c[2]

display(qc.draw())

qc.barrier()

# FIX: Modern Qiskit classical conditions using context managers
with qc.if_test((c, 5)):
    qc.h(q[0])
with qc.if_test((c, 3)):
    qc.x(q[0])
with qc.if_test((c, 1)):
    qc.z(q[0])

for i in range(5):
    if i % 2 == 0: 
        qc.h(q[i])
    else: 
        qc.x(q[i])
        
b = 3*pi/11
qc.cu(2*b, 0, 0, 0, q[1], q[0])

display(qc.draw(output='mpl', reverse_bits=True))

job = AerSimulator().run(qc, shots=1000)
counts = job.result().get_counts(qc)   
print("measurement outcomes are:", counts)
print()


# --- PART 2: 2-Qubit Circuit for Unitary and Statevector Simulation ---

q = QuantumRegister(2, "q")
c = ClassicalRegister(2, "c")

qc = QuantumCircuit(q, c)

qc.h(q[0])
qc.ry(a, q[0])
qc.x(q[1])
qc.cx(q[0], q[1])

# FIX 1: Add save_unitary() to the circuit so the Aer backend populates the data key
qc.save_unitary()

import numpy as np
unitary_sim = AerSimulator(method="unitary")
job = unitary_sim.run(qc, shots=1, optimization_level=0)

# FIX 2: get_unitary() already formats the output as an Operator, do not call .data
current_unitary = np.round(job.result().get_unitary(qc), 3)

print("unitary matrix is")
print(current_unitary)
print()

# --- PART 3: Statevector Simulation ---
# Reinitialize the circuit since save_unitary changed its instructions
qc = QuantumCircuit(q, c)
qc.h(q[0])
qc.ry(a, q[0])
qc.x(q[1])
qc.cx(q[0], q[1])

# FIX 3: Add save_statevector() before running the statevector simulation
qc.save_statevector()

state_sim = AerSimulator(method="statevector")
job = state_sim.run(qc, optimization_level=0)
current_quantum_state = np.round(job.result().get_statevector(qc), 3)

print("quantum state is")
for val in current_quantum_state:
    print(str(val.real) + "+" + str(val.imag) + "j")

print()
print("All looks fine :-)")
```
