We observe another interesting quantum effect here.

We apply a Controlled-NOT operator, but the controller qubit will be affected!

### Phase Kickback
Phase Kickback is a quantum phenomenon where the phase information of a target qubit is "kicked back" to the controller qubit during a controlled gate operation (like a CNOT).

Normally, a CNOT operation changes the state of the target qubit based on the control qubit. However, if the target qubit is in a specific superposition state—specifically an eigenstate of the target operation—the CNOT doesn't just flip the target's bit value. Instead, the phase applied to the target state is transferred (kicked back) to the control qubit, altering its superposition phase without changing the target's state.

This is a fundamental mechanism in quantum computing, enabling powerful algorithms like Phase Estimation and Shor’s Algorithm, by allowing global phase information to be mapped onto qubits that are easier to measure.

Phase kickback is a universal rule that applies to any controlled gate (such as a Controlled-Z, Controlled-Phase, or any general Controlled-U gate).

As long as the target qubit is prepared in an eigenstate of that specific gate's operation, the phase will be kicked back to the control qubit. The CNOT gate is simply the most common and accessible example used to demonstrate the effect!

### Eigenstate
an eigenstate is a specific physical state of a system that remains fundamentally unchanged when a particular observable (like energy, momentum, or spin) is measured.

Essentially, if a state is not an eigenstate, a measurement or operation will typically force the system into a superposition of different states. If it is an eigenstate, the system stays "locked" in that configuration, and you simply gain information (the eigenvalue) about it.

Phase kickback specifically relies on the target qubit being in an eigenstate of the gate operation. Because it is an eigenstate, the gate doesn't flip the qubit's bit; instead, it "kicks" the phase change back to the controller qubit.


<h3> Task 1</h3>

Create a quantum circuit with two qubits, say $ q[1] $ and $ q[0] $ in the reading order of Qiskit.

We start in quantum state $ \ket{01} $:
- set the state of $ q[1] $ to $ \ket{0} $, and
- set the state of $ q[0] $ to $ \ket{1} $.

Apply Hadamard to both qubits.

Apply CNOT operator, where the controller qubit is $ q[1] $ and the target qubit is $ q[0] $.

Apply Hadamard to both qubits.

Measure the outcomes.

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit_aer import AerSimulator

# Create a quantum circuit with 2 qubits and 2 classical bits
q = QuantumRegister(2, "q")
c = ClassicalRegister(2, "c")
qc = QuantumCircuit(q, c)

# 1. Start in state |01> 
# q[1] is already |0>, flip q[0] to |1>
qc.x(0) 

# 2. Apply Hadamard to both qubits
qc.h(0)
qc.h(1)

# 3. Apply CNOT (controller q[1], target q[0])
qc.cx(1, 0)

# 4. Apply Hadamard to both qubits
qc.h(0)
qc.h(1)

# 5. Measure the outcomes
qc.measure(q, c)

# Execute the circuit
simulator = AerSimulator()
job = simulator.run(qc, shots=100)
result = job.result()
counts = result.get_counts(qc)
print(counts)
```

<h3> The effect of CNOT </h3>

The quantum state of the up qubit before CNOT:

$$
    \ket{0} \xrightarrow{H} \frac{1}{\sqrt{2}} \ket{0} + \frac{1}{\sqrt{2}} \ket{1}.
$$

The quantum state of the down qubit before CNOT:

$$
    \ket{1} \xrightarrow{H} \frac{1}{\sqrt{2}} \ket{0} - \frac{1}{\sqrt{2}} \ket{1}.
$$

The quantum state of the composite system:

$$
   \mypar{ \frac{1}{\sqrt{2}} \ket{0} + \frac{1}{\sqrt{2}} \ket{1} } 
   \otimes
   \mypar{ \frac{1}{\sqrt{2}} \ket{0} - \frac{1}{\sqrt{2}} \ket{1}  }
$$

CNOT affects when the up qubit has the value 1.

Let's rewrite the composite state as below to explicitly represent the effect of CNOT.

$$
     \frac{1}{\sqrt{2}} \ket{0} \otimes \mypar{ \frac{1}{\sqrt{2}} \ket{0} - \frac{1}{\sqrt{2}} \ket{1}  }
     +
     \frac{1}{\sqrt{2}} \ket{1} \otimes \mypar{ \frac{1}{\sqrt{2}} \ket{0} - \frac{1}{\sqrt{2}} \ket{1}  }
$$

CNOT flips the state of the down qubit.

After CNOT, we have:

$$
     \frac{1}{\sqrt{2}} \ket{0} \otimes \mypar{ \frac{1}{\sqrt{2}} \ket{0} - \frac{1}{\sqrt{2}} \ket{1}  }
     +
     \frac{1}{\sqrt{2}} \ket{1} \otimes \mypar{ \frac{1}{\sqrt{2}} \ket{1} - \frac{1}{\sqrt{2}} \ket{0}  }
$$

Remark that $\ket{0}$ and $ \ket{1} $ are swapped in the second qubit.

If we write the quantum state of the down qubit as before, the sign of $ \ket{1} $ in the up qubit should be flipped.

Thus the last equation can be equivalently written as follows:

$$
     \frac{1}{\sqrt{2}} \ket{0} \otimes \mypar{ \frac{1}{\sqrt{2}} \ket{0} - \frac{1}{\sqrt{2}} \ket{1}  }
     -
     \frac{1}{\sqrt{2}} \ket{1} \otimes \mypar{ \frac{1}{\sqrt{2}} \ket{0} - \frac{1}{\sqrt{2}} \ket{1}  }
$$

Before CNOT operator, the sign of $ \ket{1} $ in the up qubit is positive. 
    
After CNOT operator, its sign changes to negative.

This is called <b>phase kickback</b>.

<h3> After CNOT </h3>

It is easy to see from the last expression, that the quantum states of the qubits are separable (no correlation):

$$
   \mypar{ \frac{1}{\sqrt{2}} \ket{0} - \frac{1}{\sqrt{2}} \ket{1} } 
   \otimes
   \mypar{ \frac{1}{\sqrt{2}} \ket{0} - \frac{1}{\sqrt{2}} \ket{1}  }
$$

If we apply Hadamard to each qubit, both qubits evolve to state $ \ket{1} $.

The final state is $ \ket{11} $.

<h3> Task 2 </h3>

Create a circuit  with 7 qubits, say $ q[6],\ldots,q[0] $ in the reading order of Qiskit.

Set the states of the top six qubits to $ \ket{0} $.

Set the state of the bottom qubit to $ \ket{1} $.

Apply Hadamard operators to all qubits.

Apply CNOT operator ($q[1]$,$q[0]$) 
<br>
Apply CNOT operator ($q[4]$,$q[0]$) 
<br>
Apply CNOT operator ($q[5]$,$q[0]$) 

Apply Hadamard operators to all qubits.

Measure all qubits. 

For each CNOT operator, is there a phase-kickback effect?

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit_aer import AerSimulator

# Create a circuit with 7 qubits and 7 classical bits
q = QuantumRegister(7, "q")
c = ClassicalRegister(7, "c")
qc = QuantumCircuit(q, c)

# 1. Set the bottom qubit q[0] to |1> (the top six are |0> by default)
qc.x(q[0])
qc.barrier()

# 2. Apply Hadamard to all qubits
for i in range(7):
    qc.h(q[i])
qc.barrier()

# 3. Apply CNOT operators with q[0] as the target
qc.cx(q[1], q[0])
qc.cx(q[4], q[0])
qc.cx(q[5], q[0])
qc.barrier()

# 4. Apply Hadamard to all qubits again
for i in range(7):
    qc.h(q[i])
qc.barrier()

# 5. Measure all qubits
qc.measure(q, c)

# Draw the circuit for visual confirmation
# display(qc.draw(output='mpl', reverse_bits=True))

# Execute the circuit
job = AerSimulator().run(qc, shots=100)
counts = job.result().get_counts(qc)
print(counts)

# Assuming you have already run the circuit and have the 'counts' dictionary
# counts = job.result().get_counts(qc)

# Qiskit reading order is right-to-left: q6 q5 q4 q3 q2 q1 q0
for bitstring in counts.keys():
    print(f"Measurement result (q6...q0): {bitstring}\n")
    
    # The control qubits we want to check
    control_qubits = [1, 4, 5]
    
    for q_idx in control_qubits:
        # Calculate the string index for q[i]
        # Since q0 is at the far right (index 6), q[i] is at index (6 - i)
        string_index = 6 - q_idx
        measured_value = bitstring[string_index]
        
        if measured_value == '1':
            print(f"q[{q_idx}]: Phase kickback CONFIRMED (measured as 1)")
        else:
            print(f"q[{q_idx}]: No phase kickback (measured as 0)")
            
    # For contrast, check a qubit that wasn't a control (e.g., q2)
    print(f"---\nq[2]: Was not a control, measured as {bitstring[6 - 2]}")
```
