_The overall probability must be 1 when we observe a quantum system._

For example, the following vectors <u>cannot</u> be a valid quantum state:

$$
    \myvector{ \dfrac{1}{2} \\ \dfrac{1}{2} } 
    \mbox{ and }
    \myvector{ \dfrac{\sqrt{3}}{2} \\ \dfrac{1}{\sqrt{2}} }.
$$

For the first vector, the probabilities of observing the states $\ket{0} $ and $ \ket{1} $ are $ \dfrac{1}{4} $. 

So, the overall probability of getting a result is $ \dfrac{1}{4} + \dfrac{1}{4} = \dfrac{1}{2} $, which is less than 1.

For the second vector, the probabilities of observing the states $\ket{0} $ and $ \ket{1} $ are respectively $ \dfrac{3}{4} $ and $ \dfrac{1}{2} $. 

So, the overall probability of getting a result is $ \dfrac{3}{4} + \dfrac{1}{2} = \dfrac{5}{4} $, which is greater than 1.

<font color="blue"><b>The summation of amplitude squares must be 1 for a valid quantum state.</b></font>

<font color="blue"><b>More formally, a quantum state can be represented by a vector having length 1, and vice versa.</b></font>

<i>The summation of amplitude squares gives the square of the length of vector.

But, this summation is 1, and its square root is also 1. So, we can use the term <u>length</u> in the definition.</i>

<i> <b>Technical notes:</b> We represent a quantum state as $ \ket{u} $ instead of $ u $. Remember the relation between the length and dot product: $ \norm{u} = \sqrt{\dot{u}{u}} $. 
    
In quantum computation, we use inner product instead of dot product, which is defined on complex numbers. By using bra-ket notation, $ \norm{ \ket{u} } = \sqrt{ \braket{u}{u} } = 1 $, or equivalently $ \braket{u}{u} = 1 $, where $ \braket{u}{u} $ is a short form of $ \bra{u}\ket{u} $. For real-valued vectors, $ \braket{v}{v} = \dot{v}{v}  $.
</i>

```python
#
# your code is here
#

import math
from random import randrange

def is_valid_quantum_state(state):
    """
    Checks if a given vector is a valid quantum state by verifying 
    if the sum of the squares of its amplitudes is close to 1.
    """
    # Sum the squares of the magnitudes
    sum_of_squares = sum(abs(amplitude)**2 for amplitude in state)
    
    # Use math.isclose to handle precision issues (e.g., 0.9999999999999998)
    return math.isclose(sum_of_squares, 1.0, rel_tol=1e-9, abs_tol=1e-9)


def random_quantum_state():
    """
    Generates a random valid 2-element quantum state vector.
    """
    # Pick random non-zero integers for the amplitudes
    a = randrange(-100, 100)
    b = randrange(-100, 100)
    
    # Prevent a zero vector (division by zero)
    while a == 0 and b == 0:
        a = randrange(-100, 100)
        b = randrange(-100, 100)
        
    # Calculate the normalization factor (length of the vector)
    magnitude = math.sqrt(a**2 + b**2)
    
    # Normalize the vector so that the sum of squares equals 1
    quantum_state = [a / magnitude, b / magnitude]
    return quantum_state


# --- Execution Loop (Repeat 10 times) ---

# Define the Hadamard operator matrix components: 1/sqrt(2) * [[1, 1], [1, -1]]
h_factor = 1 / math.sqrt(2)

print("Starting 10 iterations of Quantum State validation...\n")

for i in range(1, 11):
    print(f"--- Iteration {i} ---")
    
    # 1. Randomly pick a quantum state
    state = random_quantum_state()
    print(f"Picked State: [{state[0]:.4f}, {state[1]:.4f}]")
    
    # 2. Check whether the picked quantum state is valid
    is_initial_valid = is_valid_quantum_state(state)
    print(f"Is initial state valid? {is_initial_valid}")
    
    # 3. Multiply Hadamard operator with the randomly created quantum state
    # Matrix multiplication: 
    # new_x = 1/sqrt(2) * (x + y)
    # new_y = 1/sqrt(2) * (x - y)
    hadamard_state = [
        h_factor * (state[0] + state[1]),
        h_factor * (state[0] - state[1])
    ]
    print(f"After Hadamard: [{hadamard_state[0]:.4f}, {hadamard_state[1]:.4f}]")
    
    # 4. Check whether the quantum state in result is valid
    is_hadamard_valid = is_valid_quantum_state(hadamard_state)
    print(f"Is post-Hadamard state valid? {is_hadamard_valid}\n")
```
